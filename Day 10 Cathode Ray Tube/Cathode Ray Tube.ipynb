```python
#to be tried using a dataframe with index maintained whthen just sum the intervals
import pandas as pd
import math

workdir = 'C:\\Users\\ed.deane\\Python\\AdventOfCode\\Day 10 Cathode Ray Tube\\'

df= pd.read_csv(workdir+'input.txt', sep=' ',header=None)

df[1]=df[1].fillna(0)
```

```python
def round_down_to_nearest_even_number(num):
    return math.floor(num / 2) * 2

def collect(num,cycle):
    return num*cycle

cycle = 1
total = 1
signal_strength = 0
new_cycle = True
image = '#'

for index, row in df.iterrows():
    
    #Initialise cycle    
    if (row[0])=='noop':
        cycle+=1
    else:
        cycle+=2
    cycle_rounded = round_down_to_nearest_even_number(cycle)
    print ('Starting cycle', cycle, cycle_rounded, row[1], total)
    if (abs((cycle+20)%40) - row[1] < 3):
        image = image+'#'
    else:
        image = image+'.'
    
    # the other where the increment is an addr and takes it past the end of the cycle
    # in which case collect then apply the total and start the loop on 1
    
    # in which case collect then apply the existing total and start the loop on 1
    if (new_cycle and cycle_rounded != cycle and (cycle_rounded == 20 or (cycle_rounded+20)%40 == 0)):
        print ('collect excluding the new total',total)
        signal_strength = signal_strength+(total*cycle_rounded)
        image = image+'#\n'
    
    # two scenarios, one where the increment takes the loop to end of the cycle
    # in which case apply the total and collect and start the loop on 0
    
    total = total + row[1]
    new_cycle=True
    
    if (cycle_rounded == cycle and (cycle_rounded == 20 or (cycle_rounded+20)%40 == 0)):
        print ('collect including the new total',total)
        signal_strength = signal_strength+(total*cycle_rounded)
        new_cycle = False
        image = image+'#'+'\n'
        
        
signal_strength
```
