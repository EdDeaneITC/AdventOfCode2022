```python
import pandas as pd
import networkx as nx

def get_value(p):
    return df.loc[p[0],p[1]]
    
def get_edges(p): 
    
    # get a list of all the steps around the position 
    next_steps = [(p[0],p[1]-1),(p[0]-1,p[1]),(p[0]+1,p[1]),(p[0],p[1]+1)]

    
    # screen out where the next steps falls off the edge of the map
    next_steps = [x for x in next_steps if (x[0] >= 0 and x[0] < len(df) and x[1] >= 0 and x[1] < len(df.columns))]

    # for each next step, filter out to just include those one letter higher or less
    next_steps = [y for y in next_steps if 
                  ((get_value(p) == 'S')
                    or 
                  (ord(get_value(y)) <= ord(get_value(p))+1
                   and 
                   get_value(y) not in ('S','E')))]
    return next_steps

workdir = 'C:\\Users\\ed.deane\\OneDrive - Sabio Ltd\\Documents\\Customers\\IKEA\\Knowledge\\Python\\AdventOfCode\\Day 12 Hill Climbing Algorithm\\'

df= pd.read_csv(workdir+'input.txt', header=None)

# For each line of the file, create a row and column value corresponding to that in the file
for index, row in df.iterrows():
    j=0
    for i in str(row[0]):
        df.at[index,j]=i
        j+=1
        
# get the start and end position
start = df.where(df=='S').dropna(how='all').dropna(axis=1)
start = (list(start.index)[0], start.columns[0])
end = df.where(df=='E').dropna(how='all').dropna(axis=1)
end = (list(end.index)[0], end.columns[0]) 
```

```python
# Go through each field and initialise the graph with its row,column as a tuple (row,column)
# An edge is created where the adjoining field (nsew) is either the same value or one more
G = nx.DiGraph()

# Note in this sense the index is what we are interpretting as the row of the df and j is the column
for index, row in df.iterrows():
    for j in range(len(df.columns)):
        print ('Adding node', (index,j) ,get_value((index, j)))
        G.add_node((index,j))
        for k in (get_edges((index, j))):
            print (chr(9),'Adding edge from',(index,j), 'to',k, get_value(k))
            G.add_edge((index,j), k)
```

```python
# add final edge from z to E
def get_end_edges(p): 
    
    # get a list of all the steps around the position 
    next_steps = [(p[0],p[1]-1),(p[0]-1,p[1]),(p[0]+1,p[1]),(p[0],p[1]+1)]
    print (next_steps)
    next_steps = [z for z in next_steps if (get_value(z) == 'z')]
    return next_steps
    
for end_edges in get_end_edges(end):
    print('Add final edge', end_edges, end)
    G.add_edge(end_edges, end)
```

```python
for p in nx.shortest_path(G, source=start, target=end):
    print(p, get_value(p))
    
len(nx.shortest_path(G, source=start, target=end))-1
```

```python
# loop through all nodes where the value is a
lowest = 10000000
for n in G.nodes:

# Get shortest path for each and store if lower than the lowest yet
    if  get_value(n) == 'a':
        
        try:
            dist = len(nx.shortest_path(G, source=n, target=end))-1
        except:
            dist = dist;
            
        if dist < lowest:
            lowest = dist;
            print (n, get_value(n), dist)
```
