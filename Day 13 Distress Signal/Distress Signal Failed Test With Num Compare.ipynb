```python
import pandas as pd
import re

workdir = 'C:\\Users\\ed.deane\\Python\\AdventOfCode\\Day 13 Distress Signal\\'

filepath = workdir+'input.txt'

def convert_to_int(num,width):
    num = '{message:{fill}{align}{width}}'.format(message=num,fill='0',align='<',width=width)    
    return int(num)


def calc_order(left,right):
    left_str = re.sub("[^0-9]", "", left)
    right_str = re.sub("[^0-9]", "", right)
    
    width = max(len(left_str),len(right_str))
    
    ret = False
    # If both have no numbers, compare length overall
    if (left_str == '' and right_str == ''):
        print ('Compare lengths of 2 empty packets')
        if len(left) <=  len(right):
            ret = True
    # else pad both nums out to the same length and compare
    else:    
        left_num = convert_to_int(left_str,width)
        right_num = convert_to_int(right_str,width)
        if left_num <= right_num:
            ret = True
    return ret

valid_indices = []

with open(filepath) as fp:
    line = fp.readline()
    cnt = 1
    packet_cnt = 1
    while line:
        if (packet_cnt == 1):
            left = line.strip()
        elif (packet_cnt == 2):
            right = line.strip()
        else:
            if (calc_order(left,right)):
                valid_indices.append(cnt)
                print('Adding, cnt')
            packet_cnt = 0
            cnt += 1
        print("Packet {}: {}".format(cnt, line.strip()))
        line = fp.readline()
        packet_cnt +=1
```
