```python
import pandas as pd
import re
import ast

workdir = 'C:\\Users\\ed.deane\\OneDrive - Sabio Ltd\\Documents\\Customers\\IKEA\\Knowledge\\Python\\AdventOfCode\\Day 13 Distress Signal\\'

filepath = workdir+'input.txt'

def convert_to_int(num,width):
    num = '{message:{fill}{align}{width}}'.format(message=num,fill='0',align='<',width=width)    
    return int(num)


valid_indices = []
packets = {}


with open(filepath) as fp:
    line = fp.readline()
    cnt = 1
    packet_cnt = 1
    while line:
        if (packet_cnt == 1):
            left = ast.literal_eval(line.strip())
        elif (packet_cnt == 2):
            right = ast.literal_eval(line.strip())
            packets [cnt] = {"left":left, "right":right}
        else:
            #if (calc_order(left,right)):
            #    valid_indices.append(cnt)
            #    print('Adding, cnt')
            packet_cnt = 0
            cnt += 1
        #print("Packet {}: {}".format(cnt, line.strip()))
        line = fp.readline()
        packet_cnt +=1
```

```python
def compare_elements (left_lst, right_lst):
    
    print("Comparing", left_lst, "with",right_lst)
    
    #Fn loops through two lists and returns True if left is lower, False if right is lower, None if they are the same
    
    #If not immediately comparable, i.e. the value in position 1 is another list, then both are then passed to 
    #the same function which loops through until they are compared. i.e. returning True, False or None
    
    #If they are both empty then we've reached the end of the list so return None
    if not left_lst and not right_lst:
        return None

    #If left is None exit True
    if not left_lst and right_lst:
        return True

    #If right is None exit False
    if left_lst and not right_lst:
        return False
        
    for i in range(len(left_lst)):
        #If right is None, i.e. let has more values than right then exit False
        if i > len(right_lst)-1:
            return False
        print(chr(9), "Comparing", left_lst[i], "with",right_lst[i])
        
        #If both values are integers, the lower integer should come first. 
        if isinstance(left_lst[i], int) and isinstance(right_lst[i], int):
            if left_lst[i] > right_lst[i]:
                return False
            elif left_lst[i] < right_lst[i]:
                return True
            
        else:
            #If either value is a list, convert both to lists and pass to the same function
            if isinstance(left_lst[i], int):
                left_val = [left_lst[i]]
                right_val = right_lst[i]
            elif isinstance(right_lst[i], int):
                left_val = left_lst[i]
                right_val = [right_lst[i]]
            else:
                left_val = left_lst[i]
                right_val = right_lst[i]
            
            ret = compare_elements(left_val, right_val)
            if ret is not None:
                return ret
    
    # Has right still got values left, in which case return True
    if (len(left_lst) < len(right_lst)):
        return True
    return None
```
