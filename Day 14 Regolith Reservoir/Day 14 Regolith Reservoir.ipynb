```python
import pandas as pd
import ast

# Do this long hand intially, using a df or graph to store the coordinates 
# At a later date, look to see if there are better ways to acheive it with graph based algorythms

workdir = 'C:\\Users\\ed.deane\\OneDrive - Sabio Ltd\\Documents\\Customers\\IKEA\\Knowledge\\Python\\AdventOfCode\\Day 14 Regolith Reservoir\\'

df= pd.read_csv(workdir+'input.txt', header=None, sep='|')

df_2d_map = pd.DataFrame()
sand_starter = (500,0)


def make_boundary():
    min_col = sand_starter[0]
    max_col = sand_starter[0]
    min_row = sand_starter[1]
    max_row = sand_starter[1]
    
    for index, row in df.iterrows():
        for i in row['path']:
            #print(i)
            if i[0] < min_col :
                min_col = i[0]
            if i[0] > max_col :
                max_col = i[0]
            if i[1] < min_row :
                min_row = i[1]
            if i[1] > max_row :
                max_row = i[1]

    for col in range(min_col,max_col+1):
        for rw in range(min_row, max_row+1):
            df_2d_map.at[rw,col]='.'


def plot_path(path):
    print(path)
    
def get_field(p):
    return df_2d_map.loc[p[1],p[0]]

def set_field(p,val):
    df_2d_map.at[p[1],p[0]]=val
    
def path_to_list(row):
    path_str = (row.split(' -> '))
    path = []
    for i in path_str :
        path.append(ast.literal_eval(i))
    
    return path

def draw_rock_lines(path):
    #print(path)
    for r in range(1,len(path)):
        if path[r-1][0] == path[r][0]:
            print("Move verticaly along row",path[r-1][1], "to",path[r][1])
            for i in range(min(path[r-1][1], path[r][1]),max(path[r-1][1], path[r][1])+1):
                df_2d_map.at[i,path[r-1][0]]='#'
        else:
            print("Move horizontally along column from",path[r-1][0], "to",path[r][0])
            for i in range(min(path[r-1][0],path[r][0]), max(path[r-1][0],path[r][0])+1):
                df_2d_map.at[path[r-1][1],i]='#'


                

                
df['path']= df[0].apply(path_to_list)
make_boundary()
for index, row in df.iterrows():
    draw_rock_lines(row['path'])
#add_sand()

#df_2d_map
```

```python
def get_next_sand_pos(p):
    
    print (p[0],p[1])
    #look down
    try:
        if get_field((p[0],p[1]+1)) == '.':
            return (p[0],p[1]+1)

    
        #down and left
        if get_field((p[0],p[1]+1)) != '.' and  get_field((p[0]-1,p[1]+1)) == '.':
            return (p[0]-1,p[1]+1)        
        
    
        #down and right
        if get_field((p[0],p[1]+1)) != '.' and  get_field((p[0]+1,p[1]+1)) == '.':
            return (p[0]+1,p[1]+1)
        
        #if 
    except:
        return 'End'
    return None

def drop_sand():
    sand_pos = sand_starter
    ret = ()
    while ret is not None:
        ret = (get_next_sand_pos(sand_pos))
        
        if ret == 'End':
            print ('La Fin')
            raise Exception("This is the end")

        if ret is None:
            #No further moves left
            print('setting', sand_pos)
            set_field(sand_pos,'o')
        else:
            sand_pos=ret
```
