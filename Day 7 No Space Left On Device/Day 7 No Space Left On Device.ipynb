```python
import pandas as pd
import json

workdir = 'C:\\Users\\ed.deane\\Python\\AdventOfCode\\Day 7 No Space Left On Device\\'

df= pd.read_csv(workdir+'test_input.txt', header=None)

tree = {}
current_directory = ''

def change_directory(tree, current_directory, new_directory):
    #create the new directory if it does not already exist
    if(new_directory == '/'):
        if "/" in tree:
            None
        else:
            tree['/']={'name':'/','files':{}}
    #or go up a directory to the parent
    elif(new_directory == '..'):
        new_directory=tree[current_directory].get('parent')
        #print('cd up from ', current_directory, 'to', new_directory)
    return tree, new_directory

def add_listing(tree, directory, lst):
    if (directory[-1])!='/':
        directory = directory + '/'
                
    if(lst[0]=='dir'):
        # add the directory 
        if lst[1] in tree:
            None
        else:                
            fullpath = directory+lst[1]+'/'
            tree[fullpath]={'name':fullpath,
                         'parent':directory,
                         'files':{}}
    else:
        # add the file
        tree[directory]['files'][lst[1]]=int(lst[0])
    return tree

for index, row in df.iterrows():
    lst = row[0].split()
    print (lst)
    print (tree)
    if (lst[0]=='$'):
        if (lst[1]=='cd'):
            print('changing dir')
            tree, current_directory = change_directory(tree, current_directory, lst[2])
    else:
        print('adding listing', current_directory)
        tree = add_listing(tree, current_directory, lst)
        
json.dumps(tree)
```

```python
def get_size(tree,i):
    total = sum(tree[i]['files'].values())
    # add all the subtrees who have this directory as a parent
    for l in [k for k,v in tree.items() if v.get('parent') == i]:
        #print (l)
        total = total + get_size(tree,l)
    return total

#677333

sum([get_size (tree,e) for e in tree if get_size (tree,e) <= 100000])
#get_size (tree,'dtbzzl')
```
