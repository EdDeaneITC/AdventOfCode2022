```python
import pandas as pd

workdir = 'C:\\Users\\ed.deane\\Python\\AdventOfCode\\Day 8 Treetop Tree House\\'

df= pd.read_csv(workdir+'test_input.txt', header=None, names=['col1'])
```

```python
#put into list
forest_list = []
for index, row in df.iterrows():
    result = []
    for digit in str(row['col1']):
        result.append(digit)
    forest_list.append(result)

```

```python
# set up function to return true if visible in any direction
def is_visible(p_r,p_c,p_item,p_forest_list):
    size = len(p_forest_list)
    
    #First up, deal with the edges
    if (p_r==0 or p_c==0 or p_r==size-1 or p_c==size-1):
        print ('Outer limits')
        return(True,0)
    
    # Check in all directions for visibility
    # If there is any line of sight then there is visibity
    
    r_left_visible = True
    r_right_visible = True
    c_up_visible = True
    c_down_visible = True
    
    r_left_visible_trees = 0
    r_right_visible_trees = 0
    c_up_visible_trees = 0
    c_down_visible_trees = 0
    
    #check left visibility, where row is the same    
    for i in range(p_c, 0, -1):
        val = p_forest_list[p_r][i]
        r_left_visible_trees += 1
        print ('Testing if there is visibility left',p_r,i,val)
        if (val >= p_item):
            print (val + ' is higher')
            r_left_visible = False
            break
    print (r_left_visible_trees)
    
    #check right visibility, where row is the same    
    for i in range(p_c+1, size):
        val = p_forest_list[p_r][i]
        r_right_visible_trees +=1
        print ('Testing if there is visibility right',p_r,i,val)
        if (val >= p_item):
            print (val + ' is higher')
            r_right_visible = False
            break

    print (r_right_visible_trees)
    #check up visibility, where column is the same    
    for i in range(p_r, 0, -1):
        val = p_forest_list[i][p_c]
        c_up_visible_trees += 1
        print ('Testing if there is visibility up',i,p_c,val)
        if (val >= p_item):
            print (val + ' is higher')
            c_up_visible = False
            break
            
    print (c_up_visible_trees)
    #check down visibility, where column is the same    
    for i in range(p_r+1, size):
        val = p_forest_list[i][p_c]
        c_down_visible_trees += 1
        print ('Testing if there is visibility down',i,p_c,val)
        if (val >= p_item):
            print (val + ' is higher')
            c_down_visible = False
            break
    print (c_down_visible_trees)
    
    total_visible_trees = c_up_visible_trees * c_down_visible_trees * r_left_visible_trees *r_right_visible_trees
    if (r_left_visible or r_right_visible or c_up_visible or c_down_visible):
        return True, total_visible_trees
    else:
        return False, total_visible_trees

visible_counter = 0
                 
# loop through each tree up and assess its visibility   
for r,item_row in enumerate(forest_list, start=0):
    visible = False
    for c, item in enumerate(item_row,start=0):        
        print ('*************')
        print ('Testing visibility of ',r,c,item)
        visibility = is_visible(r,c,item,forest_list)
        if(visibility[0]):
            visible_counter += 1
            print ('Visible',r,c,item)
        print ('Visible trees', visibility[1])
print (visible_counter)
```
